```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression

# 1. 数据加载
def load_data(filepath, sample_size=100000):
    try:
        df = pd.read_csv(filepath, sep='\t', on_bad_lines='skip')
        print(f"原始数据集形状: {df.shape}")
        return df.sample(min(sample_size, len(df)), random_state=42)
    except Exception as e:
        print(f"加载数据时出错: {e}")
        return None

# 文件路径
file_path = "/kaggle/input/36118at2/amazon_reviews_us_Gift_Card_v1_00.tsv"
df = load_data(file_path)

# 2. 数据清洗
def clean_data(df):
    # 选择需要的列并移除3星评价
    df = df[['review_body', 'star_rating']][df.star_rating != 3].copy()
    
    # 处理缺失值
    print(f"清洗前缺失值数量:\n{df.isnull().sum()}")
    df.dropna(inplace=True)
    print(f"清洗后数据集形状: {df.shape}")
    
    # 文本清洗
    df['cleaned_review'] = df.review_body.apply(
        lambda x: re.sub(r'<[^>]+>|\W|\d|http\S+', ' ', str(x).lower()).strip()
    )
    
    # 转换为二分类问题
    df['sentiment'] = df.star_rating.replace({1: 0, 2: 0, 4: 1, 5: 1})
    return df

df_clean = clean_data(df)

# 3. 探索性分析(EDA)
def perform_eda(df):
    # 星级分布
    plt.figure(figsize=(10, 6))
    rating_counts = df.star_rating.value_counts().sort_index()
    ax = sns.barplot(x=rating_counts.index, y=rating_counts.values)
    plt.title('星级评分分布')
    plt.xlabel('星级')
    plt.ylabel('数量')
    
    # 在柱子上添加数值
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10), textcoords='offset points')
    plt.show()
    
    # 情感分布
    plt.figure(figsize=(6, 4))
    sentiment_counts = df.sentiment.value_counts()
    plt.pie(sentiment_counts, labels=['负面(1-2星)', '正面(4-5星)'], 
            autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
    plt.title('情感分布')
    plt.show()
    
    # 评论长度分析
    df['review_length'] = df.cleaned_review.apply(lambda x: len(x.split()))
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='star_rating', y='review_length', data=df)
    plt.ylim(0, 100)  # 限制长度范围以便更好观察
    plt.title('不同星级下的评论长度分布')
    plt.show()

perform_eda(df_clean)

# 4. 数据准备与建模
def prepare_and_model(df):
    # 分割数据集
    X_train, X_test, y_train, y_test = train_test_split(
        df.cleaned_review, 
        df.sentiment, 
        test_size=0.2, 
        random_state=42,
        stratify=df.sentiment  # 保持类别比例
    )
    
    # 特征提取
    vectorizers = {
        'BoW': CountVectorizer(max_features=5000),
        'TF-IDF': TfidfVectorizer(max_features=5000)
    }
    
    results = {}
    for name, vectorizer in vectorizers.items():
        # 特征转换
        X_train_vec = vectorizer.fit_transform(X_train)
        X_test_vec = vectorizer.transform(X_test)
        
        # 训练模型
        model = LogisticRegression(max_iter=1000, random_state=42)
        model.fit(X_train_vec, y_train)
        
        # 预测与评估
        y_pred = model.predict(X_test_vec)
        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred, output_dict=True)
        
        # 存储结果
        results[name] = {
            'accuracy': accuracy,
            'report': report,
            'cm': confusion_matrix(y_test, y_pred)
        }
        
        # 打印结果
        print(f"\n{name} 模型结果:")
        print(f"准确率: {accuracy:.4f}")
        print(classification_report(y_test, y_pred))
        
        # 绘制混淆矩阵
        plt.figure(figsize=(6, 4))
        sns.heatmap(results[name]['cm'], annot=True, fmt='d', cmap='Blues',
                    xticklabels=['负面', '正面'], 
                    yticklabels=['负面', '正面'])
        plt.title(f'{name} 混淆矩阵')
        plt.xlabel('预测标签')
        plt.ylabel('真实标签')
        plt.show()
    
    return results

model_results = prepare_and_model(df_clean)

# 5. 结果对比
print("\n模型性能对比:")
for model_name, result in model_results.items():
    print(f"{model_name} 准确率: {result['accuracy']:.4f}")
    print(f"正面评价F1-score: {result['report']['1']['f1-score']:.4f}")
    print(f"负面评价F1-score: {result['report']['0']['f1-score']:.4f}\n")
```

```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression

# 1. Data Loading
def load_data(filepath, sample_size=100000):
    try:
        df = pd.read_csv(filepath, sep='\t', on_bad_lines='skip')
        print(f"Original dataset shape: {df.shape}")
        return df.sample(min(sample_size, len(df)), random_state=42)
    except Exception as e:
        print(f"Error loading data: {e}")
        return None

# File path
file_path = "/kaggle/input/36118at2/amazon_reviews_us_Gift_Card_v1_00.tsv"
df = load_data(file_path)

# 2. Data Cleaning
def clean_data(df):
    # Select needed columns and remove 3-star reviews
    df = df[['review_body', 'star_rating']][df.star_rating != 3].copy()
    
    # Handle missing values
    print(f"Missing values before cleaning:\n{df.isnull().sum()}")
    df.dropna(inplace=True)
    print(f"Dataset shape after cleaning: {df.shape}")
    
    # Text cleaning
    df['cleaned_review'] = df.review_body.apply(
        lambda x: re.sub(r'<[^>]+>|\W|\d|http\S+', ' ', str(x).lower()).strip()
    )
    
    # Convert to binary sentiment
    df['sentiment'] = df.star_rating.replace({1: 0, 2: 0, 4: 1, 5: 1})
    return df

df_clean = clean_data(df)

# 3. Exploratory Data Analysis (EDA)
def perform_eda(df):
    # Star rating distribution
    plt.figure(figsize=(10, 6))
    rating_counts = df.star_rating.value_counts().sort_index()
    ax = sns.barplot(x=rating_counts.index, y=rating_counts.values)
    plt.title('Star Rating Distribution')
    plt.xlabel('Star Rating')
    plt.ylabel('Count')
    
    # Add values on bars
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', 
                   (p.get_x() + p.get_width() / 2., p.get_height()),
                   ha='center', va='center', xytext=(0, 10), textcoords='offset points')
    plt.show()
    
    # Sentiment distribution
    plt.figure(figsize=(6, 4))
    sentiment_counts = df.sentiment.value_counts()
    plt.pie(sentiment_counts, labels=['Negative (1-2 stars)', 'Positive (4-5 stars)'], 
            autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
    plt.title('Sentiment Distribution')
    plt.show()
    
    # Review length analysis
    df['review_length'] = df.cleaned_review.apply(lambda x: len(x.split()))
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='star_rating', y='review_length', data=df)
    plt.ylim(0, 100)  # Limit range for better visualization
    plt.title('Review Length Distribution by Star Rating')
    plt.show()

perform_eda(df_clean)

# 4. Data Preparation and Modeling
def prepare_and_model(df):
    # Split dataset
    X_train, X_test, y_train, y_test = train_test_split(
        df.cleaned_review, 
        df.sentiment, 
        test_size=0.2, 
        random_state=42,
        stratify=df.sentiment  # Maintain class ratio
    )
    
    # Feature extraction
    vectorizers = {
        'BoW': CountVectorizer(max_features=5000),
        'TF-IDF': TfidfVectorizer(max_features=5000)
    }
    
    results = {}
    for name, vectorizer in vectorizers.items():
        # Feature transformation
        X_train_vec = vectorizer.fit_transform(X_train)
        X_test_vec = vectorizer.transform(X_test)
        
        # Train model
        model = LogisticRegression(max_iter=1000, random_state=42)
        model.fit(X_train_vec, y_train)
        
        # Predict and evaluate
        y_pred = model.predict(X_test_vec)
        accuracy = accuracy_score(y_test, y_pred)
        report = classification_report(y_test, y_pred, output_dict=True)
        
        # Store results
        results[name] = {
            'accuracy': accuracy,
            'report': report,
            'cm': confusion_matrix(y_test, y_pred)
        }
        
        # Print results
        print(f"\n{name} Model Results:")
        print(f"Accuracy: {accuracy:.4f}")
        print(classification_report(y_test, y_pred))
        
        # Plot confusion matrix
        plt.figure(figsize=(6, 4))
        sns.heatmap(results[name]['cm'], annot=True, fmt='d', cmap='Blues',
                    xticklabels=['Negative', 'Positive'], 
                    yticklabels=['Negative', 'Positive'])
        plt.title(f'{name} Confusion Matrix')
        plt.xlabel('Predicted Label')
        plt.ylabel('True Label')
        plt.show()
    
    return results

model_results = prepare_and_model(df_clean)

# 5. Results Comparison
print("\nModel Performance Comparison:")
for model_name, result in model_results.items():
    print(f"{model_name} Accuracy: {result['accuracy']:.4f}")
    print(f"Positive Class F1-score: {result['report']['1']['f1-score']:.4f}")
    print(f"Negative Class F1-score: {result['report']['0']['f1-score']:.4f}\n")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
import os

# Download necessary NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# Set the correct data path
data_path = '/kaggle/input/d6118-facebook/emotion-emotion_69k.csv'

# Check if file exists
if not os.path.exists(data_path):
    raise FileNotFoundError(f"Dataset not found at {data_path}. Please check the path.")

# Load the dataset
try:
    df = pd.read_csv(data_path)
    print("Dataset loaded successfully!")
except Exception as e:
    print(f"Error loading dataset: {e}")
    raise

# Basic dataset info
print("\nDataset shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst few rows:")
print(df.head())

# Preprocessing function for text
def preprocess_text(text):
    if pd.isna(text):
        return ""
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize
    tokens = word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words and len(word) > 2]
    return ' '.join(tokens)

# Create word clouds for different text columns
def generate_wordcloud(text, title):
    wordcloud = WordCloud(width=800, height=400, 
                         background_color='white',
                         max_words=100).generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title, fontsize=15)
    plt.axis('off')
    plt.show()

# Preprocess and visualize Situation column
if 'Situation' in df.columns:
    situations = ' '.join(df['Situation'].dropna().astype(str))
    preprocessed_situations = preprocess_text(situations)
    generate_wordcloud(preprocessed_situations, 'Most Common Words in Situations')
else:
    print("'Situation' column not found in dataset")

# Preprocess and visualize Emotion column
if 'emotion' in df.columns:
    emotion_counts = df['emotion'].value_counts()
    plt.figure(figsize=(12, 6))
    emotion_counts.plot(kind='bar')
    plt.title('Distribution of Emotions in Conversations', pad=20)
    plt.xlabel('Emotion')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("'emotion' column not found in dataset")

# Preprocess and visualize Dialogues
if 'empathetic_dialogues' in df.columns:
    dialogues = ' '.join(df['empathetic_dialogues'].dropna().astype(str))
    preprocessed_dialogues = preprocess_text(dialogues)
    generate_wordcloud(preprocessed_dialogues, 'Most Common Words in Dialogues')
else:
    print("'empathetic_dialogues' column not found in dataset")

# Preprocess and visualize Agent Responses
if 'labels' in df.columns:
    responses = ' '.join(df['labels'].dropna().astype(str))
    preprocessed_responses = preprocess_text(responses)
    generate_wordcloud(preprocessed_responses, 'Most Common Words in Agent Responses')
else:
    print("'labels' column not found in dataset")

# Analyze most common emotion-word associations
def get_emotion_word_associations(df, top_n=5):
    emotion_words = {}
    if 'emotion' not in df.columns or 'empathetic_dialogues' not in df.columns:
        print("Required columns not found for emotion-word analysis")
        return pd.DataFrame()
    
    for emotion in df['emotion'].unique():
        if pd.isna(emotion):
            continue
        # Get all dialogues for this emotion
        emotion_dialogues = ' '.join(df[df['emotion'] == emotion]['empathetic_dialogues'].dropna().astype(str))
        preprocessed = preprocess_text(emotion_dialogues)
        words = preprocessed.split()
        word_counts = Counter(words)
        top_words = [word for word, count in word_counts.most_common(top_n)]
        emotion_words[emotion] = top_words
    
    # Create a DataFrame for better visualization
    emotion_word_df = pd.DataFrame.from_dict(emotion_words, orient='index')
    return emotion_word_df

# Get top words by emotion
emotion_word_df = get_emotion_word_associations(df)
if not emotion_word_df.empty:
    print("\nTop words associated with each emotion:")
    print(emotion_word_df)

    # Visualize word-emotion associations
    plt.figure(figsize=(14, 10))
    emotions = emotion_word_df.index
    n_emotions = len(emotions)
    n_cols = 4
    n_rows = (n_emotions // n_cols) + (1 if n_emotions % n_cols else 0)
    
    for i, emotion in enumerate(emotions):
        words = emotion_word_df.loc[emotion].tolist()
        text = ' '.join([f"{word}" for word in words])
        wordcloud = WordCloud(width=300, height=200, background_color='white').generate(text)
        plt.subplot(n_rows, n_cols, i+1)
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.title(emotion)
        plt.axis('off')
    
    plt.tight_layout()
    plt.suptitle('Top Words by Emotion', y=1.02, fontsize=16)
    plt.show()
```
